```python
import pandas as pd

# Load only available relevant columns
trader_df = pd.read_csv("historical_data.csv", usecols=[
    "Account",           # account
    "Coin",              # symbol
    "Execution Price",   # execution price
    "Size Tokens",       # size
    "Side",              # side
    "Timestamp",         # time
    "Start Position",    # start position
    "Direction",         # event
    "Closed PnL"         # closedPnL
])
```

```python
trader_df.isnull().sum()
```

```python
trader_df['Timestamp'] = pd.to_datetime(trader_df['Timestamp'], errors='coerce')
trader_df['date'] = trader_df['Timestamp'].dt.date
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'], errors='coerce').dt.date
```

```python
# Create date → sentiment mapping
sentiment_map = dict(zip(sentiment_df['date'], sentiment_df['classification']))
print(sentiment_map)
# Add sentiment column to trader data
trader_df['sentiment'] = trader_df['date'].map(sentiment_map)
```

```python
# print(trader_df)
print(trader_df[['date', 'sentiment']].drop_duplicates().head(10))
print(trader_df['sentiment'].isna().mean())  # % of trades with no sentiment
```
